```python
north_america=df.loc[df.Region=="North America"]
latin_america=df.loc[df.Region=="Latin America"]
western_europe=df.loc[df.Region=="Western Europe"]
eastern_europe=df.loc[df.Region=="Eastern Europe"]
asia_pacific=df.loc[df.Region=="Asia-Pacific"]
africa=df.loc[df.Region=="Africa"]
```

```python
listo=[north_america,latin_america,western_europe,eastern_europe,asia_pacific, africa]
mean=pd.DataFrame()
for i in listo:
    mean = mean.append({"means": i['Happiness Score'].mean()}, ignore_index=True)

print(mean)
```

```python
above_six=pd.DataFrame()
for i in listo:
    above_six = above_six.append({"above_six": i[i["Happiness Score"]>6].shape[0]}, ignore_index=True)

print(above_six)
```

```python
combined=pd.concat([above_six,mean], axis=1)
countr=pd.DataFrame(["north_america","latin_america","western_europe","eastern_europe","asia_pacific", "africa"], columns=["Regions"])
combined=pd.concat([combined,countr], axis=1)
combined
```

```python
distance=pd.DataFrame()
for i in listo:
    max=i["Happiness Score"].max()
    min=i["Happiness Score"].min()
    distance=distance.append({"distance": (max-min)}, ignore_index=True)
print(distance)                    
```

```python
combined=pd.concat([combined,distance], axis=1)
combined
```

```python
combined.sort_values("distance")
```
